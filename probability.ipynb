# Multivariate Normal Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
#Bivariate
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])   
dist=np.random.multivariate_normal(mean,cov,500)
                                                           #we can say there is a positive  correlation between them
#mean2=np.array([5.0,6.0])
#cov2=np.array([[1.3,0.2],[0.2,1.1]])  
#dist2=np.random.multivariate_normal(mean2,cov2,500)                            #in second distribution the spread is more                                  

print(dist.shape)
print(dist)
plt.scatter(dist[:,0],dist[:,1])
#plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
